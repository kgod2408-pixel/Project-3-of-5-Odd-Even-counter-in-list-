```python
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

odd_count = 0
even_count = 0

for i in range(len(numbers)):
    
    if numbers[i] % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

print("Total Even Numbers:", even_count)
print("Total Odd Numbers:", odd_count)
```
